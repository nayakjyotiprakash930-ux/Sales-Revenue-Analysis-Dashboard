```python
import pandas as pd
import numpy as np

# 1. Simulate/Import Raw Data (Excel/CSV)
np.random.seed(42)
data_size = 1000

data = {
    'Order_Date': pd.date_range(start='2025-01-01', periods=data_size, freq='D'),
    'Product_Category': np.random.choice(['Electronics', 'Clothing', 'Home Decor', 'Sports'], data_size),
    'Product_Name': np.random.choice(['Laptop', 'T-Shirt', 'Vase', 'Football', 'Headphones', 'Jeans'], data_size),
    'Quantity': np.random.randint(1, 5, size=data_size),
    'Unit_Price': np.random.uniform(15.0, 500.0, size=data_size).round(2),
    'Region': np.random.choice(['North', 'East', 'South', 'West'], data_size)
}

df = pd.DataFrame(data)

# 2. Data Transformation (Feature Engineering)
# Total Sales/Revenue = Quantity * Unit Price
df['Revenue'] = df['Quantity'] * df['Unit_Price']
# Simulating Cost of Goods Sold (COGS) to calculate Profit
df['COGS'] = df['Revenue'] * np.random.uniform(0.4, 0.7, size=data_size)
df['Profit'] = df['Revenue'] - df['COGS']

# 3. Calculate Core KPIs for the Dashboard
total_revenue = df['Revenue'].sum()
total_units_sold = df['Quantity'].sum()
average_order_value = df['Revenue'].mean()
overall_profit_margin = (df['Profit'].sum() / total_revenue) * 100

print(f"--- Core Business KPIs ---")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Units Sold: {total_units_sold:,}")
print(f"Average Order Value: ${average_order_value:,.2f}")
print(f"Overall Profit Margin: {overall_profit_margin:.2f}%\n")

# 4. Export clean data for Dashboard Tool
df.to_csv('processed_sales_data.csv', index=False)
print("Data cleaned and exported as 'processed_sales_data.csv'")
```
